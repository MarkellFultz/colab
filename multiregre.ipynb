```python
import pandas as pd
url='https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data2.csv'
data=pd.read_csv(url)
data['EducationLevel']=data['EducationLevel'].map({"高中以下":0,"大學":1,"碩士以上":2})
data
```

```python
#one hot encoding轉換單一欄位
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder()
OHE.fit(data[['City']])
city_encoded=OHE.transform(data[['City']]).toarray()
print(city_encoded)
data[["CityA","CityB","CityC",]]=city_encoded
```

```python
data=data.drop(["City","CityC"],axis=1)
data
```

```python
from sklearn.model_selection import train_test_split


x=data[["YearsExperience","EducationLevel","CityA","CityB"]]
y=data["Salary"]
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=87)#分割兩成資料做測試集 變數random_state可固定 test_size決定分割大小
X_train=X_train.to_numpy()
X_test=X_test.to_numpy()
Y_train=Y_train.to_numpy()
Y_test=Y_test.to_numpy()
```

```python
def compute_cost(x,y,b1,b2):
  pre_y=(x*b1).sum(axis=1)+b2
  cost=((y-pre_y)**2).mean()
  return cost
```

```python
#平方法
import numpy as np
b1=np.array([0,2,2,4])
b2=0
compute_cost(X_train,Y_train,b1,b2)
```

```python
#Gradient
#斜率計算(真實y-pred_y)平方=(y-(w1x1+w2x2...wnxn+b2))平方偏微
#         EX:對w1 得到 2*(-x1)*(y-(w1x1+w2x2...wnxn+b2))=2*(x1)*(pred_y-真實y)
#                ↑ ↑
# 實作上把2交給learning rate去變化 (0-(w1+0+0+0+0))

#除b2外因為b2沒有函數項目因此(pred_y-真實y)
pre_y=(X_train*b1).sum(axis=1)+b2
b2_grediant=(pre_y-Y_train).mean()
'''
x1=X_train[:,0]
w1_gredient=(X_train[:,0]*(pre_y-Y_train)).mean()
w2_gredient=(X_train[:,1]*(pre_y-Y_train)).mean()
w3_gredient=(X_train[:,2]*(pre_y-Y_train)).mean()
w4_gredient=(X_train[:,3]*(pre_y-Y_train)).mean()
w1_gredient,w2_gredient,w3_gredient,w4_gredient'''
#計算透過迴圈得到每一個變數斜率
def compute_gradient(x,y,b,b2):


  pre_y=(x*b).sum(axis=1)+b2
  feature=x.shape[1]
  w_gradient=np.zeros(feature)
  b2_gradiant=(pre_y-y).mean()
  for i in range(feature):
    w_gradient[i]=(x[:,i]*(pre_y-y)).mean()
  return w_gradient,b2_gradiant
```

```python

def gradient_descent(x,y,b1,b2,learning_rate,cost,coumpute_gradient,run_itter,p_itter):

  b1=b1
  b2=b2
  cost_history=[]
  b1_history=[]
  b2_history=[]
  p_itter=p_itter
  for i in range(run_itter):
    b1_gradient,b2_gradient=compute_gradient(x,y,b1,b2)
    b1=b1-b1_gradient*learning_rate
    b2=b2-b2_gradient*learning_rate
    costx=cost(x,y,b1,b2)
    b1_history.append(b1)
    b2_history.append(b2)
    cost_history.append(costx)

    if(i%p_itter==0):
      print(f'第{i:5}次:{costx}, b1:{b1}, b2:{b2},b1_gradient:{b1_gradient}, b2_gradient:{b2_gradient}')
  return b1,b2,cost_history,b1_history,b2_history
```

```python

learning_rate=1.0e-3
b = np.array([1, 2, 2, 4])  # ✅ 正確寫法：b 為 numpy array
b2 = 1

# 呼叫
b1_final, b2_final, cost_history, b1_history, b2_history = gradient_descent(
    X_train, Y_train, b, b2, learning_rate, compute_cost, compute_gradient, 20000, 1000)

```
