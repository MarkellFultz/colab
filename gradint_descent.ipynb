```python
import pandas as pd
url='https://raw.githubusercontent.com/GrandmaCan/ML/main/Resgression/Salary_Data.csv'
data=pd.read_csv(url)

x=data['YearsExperience']
y=data['Salary']
```

```python
'''原版
def compute_gradient(x,y,b1,b2):
  b1_gradient=(2*x*((b1*x+b2)-y)).mean()  #(y−(wx+b))2 微分chain rule 降次 2*(y−(wx+b)) *(內部) (0-(x+0))==2x*(-(y−(wx+b)))
  b2_gradient=2*((b1*x+b2)-y).mean()
  return  b1_gradient,b2_gradient
'''

#暫時未知為何不能*2
def compute_gradient(x,y,b1,b2):


    b1_gradient=(x*((b1*x+b2)-y)).mean()
    b2_gradient=((b1*x+b2)-y).mean()
    return  b1_gradient,b2_gradient
```

```python
def cost(x,y,b1,b2):
  pre_value=x*b1+b2
  cost=(y-pre_value)**2
  cost=cost.sum()/len(x)

  return cost
```

```python

''' 觀看gradient變化
b1_gradient,b2_gradient=compute_gradient(x,y,b1,b2)
org=cost(x,y,b1,b2)
print(org)

b1=b1-b1_gradient*learning_rate
b2=b2-b2_gradient*learning_rate
print(b1)
print(b2)
'''
def gradient_descent(x,y,b1,b2,learning_rate,cost,coumpute_gradient,run_itter,p_itter):

  b1=b1
  b2=b2
  cost_history=[]
  b1_history=[]
  b2_history=[]
  p_itter=p_itter
  for i in range(run_itter):
    b1_gradient,b2_gradient=compute_gradient(x,y,b1,b2)
    b1=b1-b1_gradient*learning_rate
    b2=b2-b2_gradient*learning_rate
    costx=cost(x,y,b1,b2)
    b1_history.append(b1)
    b2_history.append(b2)
    cost_history.append(costx)

    if(i%p_itter==0):
      print(f'第{i:5}次:{costx:2f}, b1:{b1:2f}, b2:{b2:2f},b1_gradient:{b1_gradient:.2f}, b2_gradient:{b2_gradient:.2f}')
  return b1,b2,cost_history,b1_history,b2_history
```

```python
b1=0
b2=0
learning_rate=1.0e-3
b1,b2,cost_history,b1_history,b2_history=gradient_descent(x,y,b1,b2,learning_rate,cost,compute_gradient,20000,1000)
```

```python
#COST曲線
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.arange(0,20000),cost_history[:20000])
plt.title('Cost History')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.show()

print(f'最終 b1={b1:.2f}, b2={b2:.2f}')

```
